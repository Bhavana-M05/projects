```python
import pandas as pd
import matplotlib.pyplot as plt
age_over_60 = data[data['Age'] > 60]
plt.figure(figsize=(20, 7))
plt.hist(age_over_60['NumberOfPurchases'], bins=10, color='Red', edgecolor='Green')
plt.title('Histogram of Number of Purchases for Customers Aged Over 60')
plt.xlabel('Number of Purchases')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='-', alpha=0.5)
plt.show()
```

```python
product_purchases = data.groupby('ProductCategory')['PurchaseStatus'].sum().sort_values(ascending=False)
plt.figure(figsize=(3,4))
plt.pie(product_purchases, labels=product_purchases.index, autopct='%1.1f%%')
plt.title('product purchases by category')
plt.axis('equal')
plt.show()
```
